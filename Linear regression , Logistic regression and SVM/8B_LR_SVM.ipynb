```python
# applying Logistic Regression on data
from sklearn.linear_model import SGDClassifier
clf_1 = SGDClassifier(loss = 'log', learning_rate = 'optimal', random_state=15, penalty='l2', tol=1e-3, verbose = 2)
clf_1.fit(X, Y)
```

```python
clf_1.coef_ , clf_1.intercept_
```

```python
'''Feature 2 and 3 are more important because of more weights being attached to them.'''
```

```python
col = data.iloc[:,:3].columns   # feature names
```

```python
# Forward Feature selection
from sklearn.feature_selection import SequentialFeatureSelector
sfs_forward = SequentialFeatureSelector(clf_1, n_features_to_select=2, direction='forward').fit(X, Y)
```

```python
print("Features selected by forward sequential selection: "f"{col[sfs_forward.get_support()]}")
```

```python
As per FFS, feature 2 and 3 are most important.
```

```python
# applying SVC on data
clf_2 = SGDClassifier(loss = 'hinge', learning_rate = 'optimal', random_state=15, penalty='l2', tol=1e-3, verbose = 2)
clf_2.fit(X, Y)
```

```python
clf_2.coef_ , clf_2.intercept_
```

```python
sfs_forward = SequentialFeatureSelector(clf_2, n_features_to_select=2, direction='forward').fit(X, Y)
```

```python
print("Features selected by forward sequential selection: "f"{col[sfs_forward.get_support()]}")
```

```python
from above, features 1 and 3 are important
```

After Standardization

```python
scaler = StandardScaler()
X_tr = scaler.fit_transform(X)
data_tr = np.concatenate((X_tr, np.array([Y]).T), axis=1)
data_tr = pd.DataFrame(data_tr, columns = ['f1', 'f2', 'f3', 'y'])
data_tr
```

```python
data_tr.corr()['y']
```

```python
data_tr.std()
```

```python
clf_3 = SGDClassifier(loss = 'log', learning_rate = 'optimal', random_state=15, penalty='l2', tol=1e-3, verbose = 2)
clf_3.fit(X_tr, Y)
```

```python
clf_3.coef_
```

```python
# applying forward feature selection
sfs_forward = SequentialFeatureSelector(clf_3, n_features_to_select=2, direction='forward').fit(X_tr, Y)
```

```python
print("Features selected by forward sequential selection: "f"{col[sfs_forward.get_support()]}")
```

```python
As per the magnitude of weights attached to the features, features 2 and 3 are the most important. This is suggested by forward 
feature selection as well.
```

```python
# applying SVC on data
clf_4 = SGDClassifier(loss = 'hinge', learning_rate = 'optimal', random_state=15, penalty='l2', tol=1e-3, verbose = 2)
clf_4.fit(X_tr, Y)
```

```python
sfs_forward = SequentialFeatureSelector(clf_4, n_features_to_select=2, direction='forward').fit(X_tr, Y)
```

```python
print("Features selected by forward sequential selection: "f"{col[sfs_forward.get_support()]}")
```
